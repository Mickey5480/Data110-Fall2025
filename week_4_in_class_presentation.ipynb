```python
x=['a','b','c','d']
y= [10,8,12,16]
z=[11,7,10,19]
plt.bar(x,y, label='week 1')
plt.bar(x,z,bottom=y, label ='week 2')
plt.legend()
plt.show()
```

```python
x=['a','b','c','d']

x_index=np.arange(len(x))
plt.bar(x_index+0.125,y,width=0.25, label='week 1')
plt.bar(x_index-0.125,z,width=0.25, label ='week 2')
```

```python
x=['a','b','c','d']

x_index=np.arange(len(x))
plt.bar(x_index+0.125,y,width=0.25, label='week 1')
plt.bar(x_index-0.125,z,width=0.25, label ='week 2')
plt.xticks(x_index,x)
plt.legend()
plt.show()
```
