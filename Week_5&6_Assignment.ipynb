```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

pclass: shows if the passenger was in 1st, 2nd, or 3rd class. Higher classes usually had better cabins and more resources.

sex: tells whether the passenger was male or female. This mattered because women and children were often rescued first.

age: shows how old the passenger was and helps us see if younger or older people were more likely to survive.

fare: the price of the ticket or higher fares often mean wealthier passengers, which could affect survival chances.

embark_town: the city where the passenger got on the ship (Cherbourg, Queenstown, or Southampton).

sibsp: how many brothers, sisters, or spouses the passenger had on board.

parch: how many parents or children the passenger had on board.

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=titanic, x="fare", bins=40, kde=True, color="skyblue")
plt.title("Distribution of Passenger Fares")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

The graph is right-skewed
The chart shows that most passengers were lower or middle class, and only a small number were wealthy.

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=titanic, x="fare", hue="survived", fill=True, common_norm=False, alpha=0.5)
plt.title("Fare Distribution by Survival Status")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

When we compare the fare distributions, we see that passengers who survived generally paid higher fares than those who did not.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=titanic, x="pclass", y="fare", hue="sex")
plt.title("Fare by Passenger Class and Sex")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

Both men and women show the same pattern. People in 1st class paid much higher fares than those in 2nd or 3rd class.

```python
plt.figure(figsize=(6,5))
sns.violinplot(data=titanic, x="sex", y="age", inner="quartile", palette="pastel")
plt.title("Age Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

This plot shows that most passangers are beteen 20-40 years old for both male and female. Both sexs are very similar.

```python
plt.figure(figsize=(6,5))
sns.violinplot(data=titanic, x="sex", y="age", inner=None, color="lightgray")
sns.stripplot(data=titanic, x="sex", y="age", size=3, jitter=True, alpha=0.6)
plt.title("Age by Sex with Individual Observations")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.show()
```

In this plot you can see each passangers age as a point. This helps define the violine plot a little bit more by showing clusters of ages and outliers too.
