```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from joypy import joyplot
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Mickey5480/Data110-21843--fall25/refs/heads/main/dataset/lincoln_temps.csv")
df.head()
```

```python
date_value = 16801
converted = pd.to_datetime('1970-01-01') + pd.to_timedelta(date_value, unit= 'D')
print(converted)
```

```python
sns.boxenplot(x='month', y='mean_temp', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxenplot(x='month', y='mean_temp', data=df, color='skyblue')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxenplot(x='month_long', y='mean_temp', data=df, color='skyblue')
plt.xticks('rotation=90')
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
sns.violinplot(x='month', y='mean_temp', data=df)
#plt.xticks('rotation=90')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='month', y='mean_temp', data=df, inner=None)
#plt.xticks('rotation=90')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='month', y='mean_temp', data=df,)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='month', y='mean_temp', data=df,color='black', jitter = 0.15)
plt.xticks('rotation=90', color='gray',fontsize='12')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='month', y='mean_temp', data=df ,color='black', jitter = 0.2)
sns.violinplot(x='month', y='mean_temp', data=df, inner=None)
plt.xticks('rotation=90', color='gray',fontsize='10')
plt.show()
```

```python
joyplot(
    data=df,
    by='month',
    column='mean_temp',

)
```

```python
joyplot(
    data=df,
    by='month',
    column='mean_temp',
    figsize=(10,10)
)
```

```python
joyplot(
    data=df,
    by='month',
    column='mean_temp',
    figsize=(10,10),
)
plt.show
```

```python
fig, ax = joyplot(
    data = df,
    by = 'month',
    column = 'mean_temp',
    figsize = (10,10),
    colormap=plt.cm.magma
)
plt.show()
```
