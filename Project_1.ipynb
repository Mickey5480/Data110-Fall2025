```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://drive.google.com/uc?export=download&id=1E4irLJ-IRbuIOpD2imtg5LnsurVAfPHU"
df = pd.read_csv(url)
print(df.head())
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Statistics.Flights.Delayed"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Total Delayed Flights Across Airports")
plt.xlabel("Number of Delayed Flights")
plt.ylabel("Frequency of Airports")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Statistics.Flights.Total"],
                y=df["Statistics.Flights.Delayed"],
                color="purple", alpha=0.7)
plt.title("Relationship Between Total Flights and Delayed Flights")
plt.xlabel("Total Flights")
plt.ylabel("Delayed Flights")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df["Statistics.Flights.Total"],
    y=df["Statistics.Flights.Delayed"],
    hue=df["Airport.Code"],  # optional: color by airport
    palette="cool",
    alpha=0.8,
    s=80
)

# Add titles and labels
plt.title("Relationship Between Total Flights and Delayed Flights (Airports)", fontsize=14)
plt.xlabel("Total Flights", fontsize=12)
plt.ylabel("Delayed Flights", fontsize=12)
plt.legend(title="Airport Code", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
delay_types = [
    "Statistics.Minutes Delayed.Carrier",
    "Statistics.Minutes Delayed.Weather",
    "Statistics.Minutes Delayed.Late Aircraft",
    "Statistics.Minutes Delayed.National Aviation System",
    "Statistics.Minutes Delayed.Security"
]

avg_delays = df[delay_types].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_delays.values, y=avg_delays.index, palette="viridis")
plt.title("Average Delay Minutes by Cause (All Airports)")
plt.xlabel("Average Minutes Delayed")
plt.ylabel("Delay Type")
plt.show()
```
