```python
import numpy as np
import matplotlib.pyplot as plt
print(plt.style.available)
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
x = np.linspace(-2, 4, 300)
y = x**2 - 2*x + 1
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-2, 2, 400)
f = x**2
g = x**3
h = x**4
i = x**5
plt.plot(x, f, "b-", label=r"$f(x)=x^2$")
plt.plot(x, g, color="red", label=r"$g(x)=x^3$")
plt.plot(x, h, color=(0.1, 0.7, 0.2), label=r"$h(x)=x^4$")
plt.plot(x, i, color="#FF00FF", label=r"$i(x)=x^5$")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = np.linspace(-2, 2, 400)
f = x**2
g = x**3
h = x**4
i = x**5
plt.plot(x, f, "b-", label=r"$f(x)=x^2$")
plt.plot(x, g, color="red", linestyle="--", label=r"$g(x)=x^3$")
plt.plot(x, h, color=(0.1, 0.7, 0.2), linestyle=":", label=r"$h(x)=x^4$")
plt.plot(x, i, color="#FF00FF", linestyle="-.", label=r"$i(x)=x^5$")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x = np.linspace(-10, 10, 400)
f = 1 / (x**2 + 1)
plt.plot(x, f, label=r"$f(x)=\frac{1}{x^2+1}$")
plt.title("Function with Custom Axes Limits")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-0.1, 1.2)   # custom y-limits
plt.xlim(-5, 5)       # custom x-limits
plt.legend()
plt.show()
plt.plot(f, x, label="Reversed Axes")
plt.title("Reversed x and y Axes")
plt.xlabel("f(x)")
plt.ylabel("x")
plt.axis("tight")   # make the axes fit tightly
plt.legend()
plt.show()
plt.plot(x, f, color="red")
plt.title("Axis Equal Example")
plt.axis("equal")
plt.show()
```

```python
x = np.linspace(0, 10, 400)
f = np.sqrt(x)
g = np.cbrt(x)
plt.plot(x, f, label=r"$f(x)=\sqrt{x}$", color="blue")
plt.plot(x, g, label=r"$g(x)=x^{1/3}$", color="red")
plt.title("Square Root vs Cube Root")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x, y, 'o', label="Scatter with plt.plot()")
plt.title("Scatter Plot using plt.plot()")
plt.legend()
plt.show()
plt.scatter(x, y, color="green", label="Scatter with plt.scatter()")
plt.title("Scatter Plot using plt.scatter()")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)
y = x * np.sin(x)
plt.plot(x, y, linestyle="-", color="lightgray", label="Line")
plt.scatter(x, y,s=80,c="blue",edgecolors="red",linewidths=.75, marker="*",
            label=r"$f(x)=x\sin(x)$")
plt.title("Customizing Scatter Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = x * np.cos(x)
plt.plot(x, y, label=r"$f(x)=x\cos(x)$", color="purple", linestyle="-")
x_points = np.arange(-2*np.pi, 2*np.pi + np.pi/4, np.pi/4)
y_points = x_points * np.cos(x_points)
plt.scatter(x_points, y_points,color="orange",
            edgecolors="black", s=80, label="Sample Points")
plt.title("Combined Plot of Line and Scatter")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
plt.style.use("ggplot") #ggplot helps make a more clear and neater gragh allowing the different styles to pop.
x = np.linspace(-10, 10, 400)
h = np.sinc(x/np.pi)
i = np.arctan(x)
j = 1 / (1 + np.exp(-x))
#Alternitive styles helps diferenciate the lines from eachother affecting the plots appearance
plt.plot(x, h, label=r"$h(x)=\frac{\sin(x)}{x}$", color="blue", linestyle="--")
plt.plot(x, i, label=r"$i(x)=\arctan(x)$", color="red", linestyle="-")
plt.plot(x, j, label=r"$j(x)=\frac{1}{1+e^{-x}}$", color="green", linestyle="-.")
plt.scatter([0, 2, -2], [np.sinc(0), np.arctan(2), 1/(1+np.exp(2))],
            color="purple", s=100, edgecolors="black", label="Highlighted Points")
plt.title("Custom Plot with Alternative Style")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="lower right")
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = x * np.cos(x)
x_points = np.arange(-2*np.pi, 2*np.pi + np.pi/4, np.pi/4)
y_points = x_points * np.cos(x_points)
# For this graph the defult style adds a blank background that doesn't help line up the data points so this grid would not be my first option.
plt.style.use("default")
plt.plot(x, y, label=r"$f(x)=x\cos(x)$", color="blue")
plt.scatter(x_points, y_points, color="red", edgecolors="black", s=70, label="Sample Points")
plt.title("Combined Plot (Default Style)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
# The bmh style works well with this grapgh becasue it helps the data points/line and the appear more clearly.
plt.style.use("bmh")
plt.plot(x, y, label=r"$f(x)=x\cos(x)$", color="green")
plt.scatter(x_points, y_points, color="orange", edgecolors="black", s=70, label="Sample Points")
plt.title("Combined Plot (Seaborn Whitegrid)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
# The seaborn-v0_8-darkgrid is not a terrible option for this graph but I don't like how the background absorbes the colors of the dataset making it a little harder to read.
plt.style.use("seaborn-v0_8-darkgrid")
plt.plot(x, y, label=r"$f(x)=x\cos(x)$", color="purple")
plt.scatter(x_points, y_points, color="yellow", edgecolors="black", s=70, label="Sample Points")
plt.title("Combined Plot (Seaborn Darkgrid)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
