```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
offenses = [
    "Aggravated Assault",
    "Burglary",
    "Larceny",
    "Motor Vehicle Theft",
    "Murder",
    "Rape",
    "Robbery"]
rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(
    offenses,
    rates_2023,
    color=["green", "pink", "blue", "Orange", "black", "purple", "red"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.ylabel("Offenses per 100,000 People (2023)")
plt.xlabel("Offense Type")
plt.title("Crime Rates per 100,000 People in 2023")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
offenses = [
    "Aggravated Assault",
    "Burglary",
    "Larceny",
    "Motor Vehicle Theft",
    "Murder",
    "Rape",
    "Robbery"]
change_2022_2023 = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]
colors =["green", "pink", "blue", "Orange", "black", "purple", "red"]
```

```python
#Task 1 shows absolute crime numbers.
#Task 2 focuses on percentage changes over time.
#Horizontal graphs like this in task 2 help identify whether crimes are rising or falling, not just how common they are.
plt.figure(figsize=(10, 6))
plt.barh(offenses, change_2022_2023, color=colors)
plt.xlabel("Change in Offenses per 100,000 People (20222023)")
plt.ylabel("Offense Type")
plt.title("Change in Crime Rates (2022 vs. 2023)")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
companies = ["Company A", "Company B", "Company C"]
products = ["Product 1", "Product 2", "Product 3"]
sales = np.array([
    [20, 34, 30],  # Company A
    [25, 32, 34],  # Company B
    [30, 35, 27]   # Company C
])
```

```python
x = np.arange(len(companies))
bar_width = 0.25
plt.figure(figsize=(10,6))
for i in range(len(products)):
    plt.bar(x + i*bar_width, sales[:, i], width=bar_width, label=products[i])
plt.xlabel("Companies")
plt.ylabel("Sales")
plt.title("Sales by Company and Product (Fake Data)")
plt.xticks(x + bar_width, companies)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
